```python
import pandas as pd

airbnb=pd.read_csv('USAirbnb.csv',quoting=3, error_bad_lines=False)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

airbnb['PropertyType']= label_encoder.fit_transform(airbnb['PropertyType'])
airbnb['RoomType']= label_encoder.fit_transform(airbnb['RoomType'])
airbnb['BedType']= label_encoder.fit_transform(airbnb['BedType'])
airbnb['Zipcode']= label_encoder.fit_transform(airbnb['Zipcode'])
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(airbnb, test_size=0.2)
```
